```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5, 11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Forma funcional
plt.plot(x, y, 'r--')
plt.xlabel('eixo X')
plt.ylabel('eixo Y')
plt.title('Titulo')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x,y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Titulo')
axes2.plot(y,x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, '8--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('titulo')
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x,y,'r--', label='x ^ 2')
axes.set_title('titulo')
axes.plot(y,x,'g--', label='x ** 1/2')
axes.legend(loc=0)
```
